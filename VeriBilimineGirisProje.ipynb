```python
import pandas as pd

# Veri setini yükleme
df = pd.read_csv('StudentPerformanceFactors.csv')
```

### Veri Seti Hikayesi

Bu veri seti, sınavlardaki öğrenci performansını etkileyen çeşitli faktörlerin kapsamlı bir özetini sunmaktadır. Ders çalışma alışkanlıkları, devam durumu, ebeveyn katılımı ve akademik başarıyı etkileyen diğer etmenlerle ilgili bilgileri içermektedir.

### Veri Seti Özellikler

- **Hours_Studied**: Haftada ders çalışmaya harcanan saat sayısı.
- **Attendance**: Derslere katılım yüzdesi.
- **Parental_Involvement**: Öğrencinin eğitimine ebeveyn katılım seviyesi (Düşük, Orta, Yüksek).
- **Access_to_Resources**: Eğitim kaynaklarına erişim durumu (Düşük, Orta, Yüksek).
- **Extracurricular_Activities**: Ders dışı etkinliklere katılım (Evet, Hayır).
- **Sleep_Hours**: Her gece ortalama uyku saati.
- **Previous_Scores**: Önceki sınavlardan alınan puanlar.
- **Motivation_Level**: Öğrencinin motivasyon seviyesi (Düşük, Orta, Yüksek).
- **Internet_Access**: İnternet erişimi durumu (Evet, Hayır).
- **Tutoring_Sessions**: Aylık alınan özel ders sayısı.
- **Family_Income**: Aile gelir seviyesi (Düşük, Orta, Yüksek).
- **Teacher_Quality**: Öğretmenlerin kalitesi (Düşük, Orta, Yüksek).
- **School_Type**: Katılınan okul türü (Devlet, Özel).
- **Peer_Influence**: Arkadaşların akademik performansa etkisi (Pozitif, Nötr, Negatif).
- **Physical_Activity**: Haftada yapılan ortalama fiziksel aktivite saati.
- **Learning_Disabilities**: Öğrenme güçlüğü olup olmadığı (Evet, Hayır).
- **Parental_Education_Level**: Ebeveynlerin en yüksek eğitim seviyesi (Lise, Üniversite, Lisansüstü).
- **Distance_from_Home**: Evden okula olan mesafe (Yakın, Orta, Uzak).
- **Gender**: Öğrencinin cinsiyeti (Erkrek, Kadın).
- **Exam_Scoe**: Final sınavı puanı.
